```python
#Address Bar feature extraction

from urllib.parse import urlparse,urlencode
import ipaddress
import re
```

```python
def getDomain(url):
  domain=urlparse(url).netloc
  if re.match(r"^www.",domain):
    domain=domain.replace("www.","")
  return domain
```

```python
print(getDomain("https://www.example.com/path/page"))
```

```python
#2 check if IP address is present in the address

def havingIP(url):
  try:
    ipaddress.ip_address(url)
    ip=1
  except:
    ip=0
  return ip
```

```python
#3 check the presence of @ in url
def haveAtSign(url):
  if "@" in url:
    at=1
  else:
    at=0
  return at
```

```python
#4 length of url
def getLength(url):
  if len(url) < 54:
    length = 0
  else:
    length = 1
  return length
```

```python
#5 depth of URl

def getDepth(url):
  s = urlparse(url).path.split('/')
  depth = 0
  for j in range(len(s)):
    if len(s[j]) != 0:
      depth = depth+1
  return depth
```

```python
#6 checking for redirection

def redirection(url):
  pos = url.rfind('//')
  if pos > 6:
    if pos > 7:
      return 1
    else:
      return 0
  else:
    return 0
```

```python
# 7.Existence of “HTTPS” Token in the Domain Part of the URL (https_Domain)
def httpDomain(url):
  domain = urlparse(url).netloc
  if 'https' in domain:
    return 1
  else:
    return 0
```

```python
shortening_services = r"bit\.ly|goo\.gl|shorte\.st|go2l\.ink|x\.co|ow\.ly|t\.co|tinyurl|tr\.im|is\.gd|cli\.gs|" \
                      r"yfrog\.com|migre\.me|ff\.im|tiny\.cc|url4\.eu|twit\.ac|su\.pr|twurl\.nl|snipurl\.com|" \
                      r"short\.to|BudURL\.com|ping\.fm|post\.ly|Just\.as|bkite\.com|snipr\.com|fic\.kr|loopt\.us|" \
                      r"doiop\.com|short\.ie|kl\.am|wp\.me|rubyurl\.com|om\.ly|to\.ly|bit\.do|t\.co|lnkd\.in|db\.tt|" \
                      r"qr\.ae|adf\.ly|goo\.gl|bitly\.com|cur\.lv|tinyurl\.com|ow\.ly|bit\.ly|ity\.im|q\.gs|is\.gd|" \
                      r"po\.st|bc\.vc|twitthis\.com|u\.to|j\.mp|buzurl\.com|cutt\.us|u\.bb|yourls\.org|x\.co|" \
                      r"prettylinkpro\.com|scrnch\.me|filoops\.info|vzturl\.com|qr\.net|1url\.com|tweez\.me|v\.gd|" \
                      r"tr\.im|link\.zip\.net"
```

```python
# 8. Checking for Shortening Services in URL (Tiny_URL)
def tinyURL(url):
    match=re.search(shortening_services,url)
    if match:
        return 1
    else:
        return 0
```

```python
# 9.Checking for Prefix or Suffix Separated by (-) in the Domain (Prefix/Suffix)
def prefixSuffix(url):
    if '-' in urlparse(url).netloc:
        return 1            # phishing
    else:
        return 0
```

```python
import requests

def web_traffic(url):
    try:
        # Extract domain from URL
        domain = url.split("//")[-1].split("/")[0]  # Removes 'https://' or 'http://'

        # Open PageRank API Endpoint
        API_KEY = "8kgscww40gw4scc8skswgosggg88o04kos88oow4"  # Replace with your API key
        api_url = f"https://openpagerank.com/api/v1.0/getPageRank?domains[]={domain}"
        headers = {"API-OPR": API_KEY}

        response = requests.get(api_url, headers=headers)
        data = response.json()

        if "response" in data and data["response"]:
            rank_str = data["response"][0].get("rank", "0")  # Default to "0" if missing

            # Ensure rank_str is not None before conversion
            if rank_str and isinstance(rank_str, str) and rank_str.isdigit():
                rank = int(rank_str)
            else:
                rank = 0  # Default rank to 0 if None or invalid

            if rank > 2:  # Rank > 2 means it's a popular site
                return 1  # Likely safe
        return 0  # Low-ranked or unknown site (suspicious)

    except Exception as e:
        print("Error fetching web traffic data:", e)
        return 0  # Assume low traffic if an error occurs

```

```python
from datetime import datetime

def domainAge(whois_data, domain_name=""):
    """
    Check if domain is suspicious based on age (<6 months)
    Returns: 1 (suspicious) or 0 (not suspicious)
    """
    try:
        # Extract creation date
        if hasattr(whois_data, 'creation_date'):  # For whois library objects
            creation_date = whois_data.creation_date
        else:  # For dictionary responses
            creation_date = whois_data.get('creation_date')

        if creation_date is None:
            return 1  # Suspicious if no creation date

        # Handle list of dates
        if isinstance(creation_date, list):
            creation_date = creation_date[0]

        # Parse string dates
        if isinstance(creation_date, str):
            creation_date = datetime.strptime(creation_date.split('+')[0].strip(), "%Y-%m-%d %H:%M:%S")

        # Calculate age in months
        age_months = (datetime.now() - creation_date).days / 30

        return 1 if age_months < 6 else 0

    except Exception as e:
        print(f"Error fetching domain age for {domain_name}: {e}")
        return 1  # Assume suspicious on error

def domainEnd(whois_data, domain_name=""):
    """
    Check if domain is suspicious based on expiration (<6 months remaining)
    Returns: 1 (suspicious) or 0 (not suspicious)
    """
    try:
        # Extract expiration date
        if hasattr(whois_data, 'expiration_date'):  # For whois library objects
            expiration_date = whois_data.expiration_date
        else:  # For dictionary responses
            expiration_date = whois_data.get('expiration_date')

        if expiration_date is None:
            return 1  # Suspicious if no expiration date

        # Handle list of dates
        if isinstance(expiration_date, list):
            expiration_date = expiration_date[0]

        # Parse string dates
        if isinstance(expiration_date, str):
            expiration_date = datetime.strptime(expiration_date.split('+')[0].strip(), "%Y-%m-%d %H:%M:%S")

        # Calculate months remaining
        months_remaining = (expiration_date - datetime.now()).days / 30

        return 1 if months_remaining < 6 else 0

    except Exception as e:
        print(f"Error fetching domain expiration for {domain_name}: {e}")
        return 1  # Assume suspicious on error
```

```python
# 15. IFrame Redirection (iFrame)
def iframe(response):
  if response == "":
      return 1
  else:
      if re.findall(r"[|]", response.text):
          return 0
      else:
          return 1
```

```python
# 16.Checks the effect of mouse over on status bar (Mouse_Over)
def mouseOver(response):
  if response == "" :
    return 1
  else:
    if re.findall("", response.text):
      return 1
    else:
      return 0
```

```python
# 17.Checks the status of the right click attribute (Right_Click)
def rightClick(response):
  if response == "":
    return 1
  else:
    if re.findall(r"event.button ?== ?2", response.text):
      return 0
    else:
      return 1
```

```python
# 18.Checks the number of forwardings (Web_Forwards)
def forwarding(response):
  if response == "":
    return 1
  else:
    if len(response.history) <= 2:
      return 0
    else:
      return 1
```

```python
#computing URL features

def featureExtraction(url,label):

  features = []
  #Address bar based features (10)
  features.append(getDomain(url))
  features.append(havingIP(url))
  features.append(haveAtSign(url))
  features.append(getLength(url))
  features.append(getDepth(url))
  features.append(redirection(url))
  features.append(httpDomain(url))
  features.append(tinyURL(url))
  features.append(prefixSuffix(url))

  #Domain based features (4)
  dns = 0
  try:
    domain_name = whois.whois(urlparse(url).netloc)
  except:
    dns = 1

  features.append(dns)
  features.append(web_traffic(url))
  features.append(1 if dns == 1 else domainAge(domain_name))
  features.append(1 if dns == 1 else domainEnd(domain_name))

  # HTML & Javascript based features (4)
  try:
    response = requests.get(url)
  except:
    response = ""
  features.append(iframe(response))
  features.append(mouseOver(response))
  features.append(rightClick(response))
  features.append(forwarding(response))
  features.append(label)

  return features
```
